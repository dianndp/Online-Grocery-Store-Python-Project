```python
# Data
data = {
    'Mid-range product': 21860860,
    'Low-range product': 10126321,
    'High-range product': 417678
}

# Create a figure and axis
fig, ax = plt.subplots()

# Custom colors
colors = ['#003D29', '#BA0239', '#FF7009']

# Create a pie chart
wedges, texts, autotexts = ax.pie(
    data.values(),
    labels=data.keys(),
    autopct='%1.1f%%',  # Display percentages on the chart
    startangle=90,
    wedgeprops={'edgecolor': '#FAF1E5'},# Add a white border to create a donut effect
    colors=colors,  # Set custom colors
    textprops={'fontsize': 14}
)

# Set aspect ratio to make it a circle (donut)
ax.set_aspect('equal')

# Add a circle in the center to make it a donut
center_circle = plt.Circle((0, 0), 0.60, fc='#FAF1E5')
fig.gca().add_artist(center_circle)

# Manually position the percentages just below the labels
for i, autotext in enumerate(autotexts):
    x, y = texts[i].get_position()
    autotext.set_position((x + 0.7, y - 0.20))  # Adjust the vertical position
    
# Manually adjust the position of the "Mid-range product" label to the right
for i, label in enumerate(texts):
    if label.get_text() == 'Mid-range product':
        x, y = label.get_position()
        label.set_position((x + 1.2, y))  # Adjust the horizontal position


# Remove the title
plt.title('')

# Export the chart as png file
plt.savefig(r'C:\Users\Admin\Desktop\Data Tasks\Data Immersion\Achievement 4\Instacart Basket Analysis\04 Analysis\Visualizations\donut_product_range.png',bbox_inches='tight',facecolor='#FAF1E5',dpi=200,transparent=True)

# Show the chart
plt.show()
```
